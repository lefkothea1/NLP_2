```python
#loading finetuned MODEL FROM DRIVE
bert_mod = ClassificationModel("bert", "/content/drive/MyDrive/NLP2/trained_models2")
```

```python
#adding negation from tutorial
from checklist.perturb import Perturb


pdata = list(nlp.pipe(data['text'])) #run cell again if u get error!!
ret = Perturb.perturb(pdata, Perturb.add_negation)
#ret.data 
#we now have a lol. inner lists: original tweet, negation of original
```

```python
#keeping only negations
neg_lst=[]
for i in range(len(ret.data)):
  neg_lst.append(ret.data[i][1])

(neg_lst)
```

```python
#geting predictions on the negations
prediction_neg, out_nr_neg = bert_mod.predict(neg_lst)
```

```python
from  checklist.editor import Editor

#making templates using {mask}
editor = Editor()

mask_pos = editor.template("I hate {mask}", nsamples=15).data
mask_neg = editor.template("I don't hate {mask}", nsamples=15).data
nation_pos = editor.template("I hate {nationality}", nsamples=15).data
nation_neg = editor.template("I don't hate {nationality}", nsamples=15).data
relig_pos = editor.template("I hate {religion}", nsamples=15).data
relig_neg = editor.template("I don't hate {religion}", nsamples=15).data

print(mask_pos)
print(mask_neg)
print(nation_pos)
print(nation_neg)
print(relig_pos)
```

```python
#geting predictions 
pred_mask_pos, out_nr_maskpos = bert_mod.predict(mask_pos)
pred_mask_neg, out_nr_maskneg = bert_mod.predict(mask_neg)
pred_nation_pos, out_nr_nationpos = bert_mod.predict(nation_pos)
pred_nation_neg, out_nr_nationneg = bert_mod.predict(nation_neg)
pred_relig_pos, out_nr_religpos = bert_mod.predict(relig_pos)
pred_relig_neg, out_nr_religneg = bert_mod.predict(relig_neg)
```
