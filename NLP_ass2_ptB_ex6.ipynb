```python
#should i install it every time?
!pip install simpletransformers 

from simpletransformers.classification import ClassificationModel, ClassificationArgs
```

```python
#loading finetuned MODEL FROM DRIVE
bert_mod = ClassificationModel("bert", "/content/drive/MyDrive/NLP2/trained_models")
```

```python
#adding negation from tutorial
from checklist.perturb import Perturb


pdata = list(nlp.pipe(data['text']))
ret = Perturb.perturb(pdata, Perturb.add_negation)
ret.data 
#we now have a lol. inner lists: original tweet, negation of original
```

```python
#keeping only negations
neg_lst=[]
for i in range(len(ret.data)):
  neg_lst.append(ret.data[i][1])

(neg_lst)
```

```python
#geting predictions on the negations
prediction_neg, out_nr_neg = bert_mod.predict(neg_lst)
```
